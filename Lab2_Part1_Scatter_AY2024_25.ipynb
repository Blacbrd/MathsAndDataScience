```python
df2 = df.copy(deep=True)

df2 = df2.dropna()

df2_filtered = df2[
    (df2["Level of study"].str.contains("undergraduate")) &
    (df2["Mode of study"] == "Full-time") &
    (df2["Category marker"] == "Sex") &
    ~(df2["Country of HE provider"] == "All")
    ]

result = df2_filtered.groupby(
    [ "HE provider", "Country of HE provider", "Category"],
    as_index=False
)["Number"].sum()

display(result)
```
